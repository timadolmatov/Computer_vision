```python
With 
A ( 
Select 
Client_id, 
Max(date)
From table
Group by client_id ) 
Select 
Client_id, 
Date, 
Case when b.Value is null then a.value else b.value end as value 
From table b 
Left Join a on b.client_id=a.client_id

# <YOUR_CODE_HERE>
```
